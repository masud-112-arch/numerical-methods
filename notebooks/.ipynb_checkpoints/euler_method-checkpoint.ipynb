# Euler’s Method for Solving Ordinary Differential Equations (ODEs)

This notebook demonstrates Euler’s method, a simple numerical technique to approximate the solution of an ODE.

We will solve the differential equation:

\[
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
\]

using Euler’s iterative scheme:

\[
y_{n+1} = y_n + h \cdot f(t_n, y_n)
\]

where \(h\) is the step size.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(f, t0, y0, h, n):
    """
    Implements Euler's method for solving ODEs.

    Parameters:
    f  : function, derivative f(t, y)
    t0 : initial time
    y0 : initial condition
    h  : step size
    n  : number of steps

    Returns:
    t_values, y_values : arrays of computed time and y values
    """
    t_values = np.zeros(n+1)
    y_values = np.zeros(n+1)

    t_values[0] = t0
    y_values[0] = y0

    for i in range(n):
        y_values[i+1] = y_values[i] + h * f(t_values[i], y_values[i])
        t_values[i+1] = t_values[i] + h

    return t_values, y_values
```

```python
# Example: dy/dt = y, y(0) = 1
f = lambda t, y: y

t0, y0 = 0, 1     # Initial condition
h = 0.1           # Step size
n = 20            # Number of steps

t_euler, y_euler = euler(f, t0, y0, h, n)

# Exact solution: y(t) = e^t
t_exact = np.linspace(t0, t0 + n*h, 100)
y_exact = np.exp(t_exact)
```

```python
plt.figure(figsize=(8,5))
plt.plot(t_exact, y_exact, label="Exact solution $e^t$", color="blue")
plt.plot(t_euler, y_euler, "o--", label="Euler’s method", color="red")

plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Euler’s Method vs Exact Solution")
plt.legend()
plt.grid(True)
plt.show()
```

## Observations

- Euler’s method provides a simple but approximate solution.
- The accuracy depends heavily on the step size (h):
  - Smaller h → more accurate but more computation.
  - Larger h → faster but less accurate.
- In this example (`dy/dt = y`), Euler’s method underestimates the true exponential growth
