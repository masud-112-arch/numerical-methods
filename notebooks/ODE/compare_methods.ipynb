# Comparison of Numerical Methods for ODEs

This notebook compares four numerical methods we have implemented:

1. Euler Method
2. Heun's Method
3. Adams–Bashforth Method
4. Runge–Kutta 4 (RK4) Method

We solve the simple exponential decay problem:

$$
\frac{dy}{dt} = -\lambda y, \quad y(0) = 1
$$

and compare the numerical solutions with the exact solution:

$$
y(t) = e^{-\lambda t}
$$

We also analyze errors for different step sizes.

```python
import numpy as np

def ode_solver(func, y0, t0, t_end, h, method="rk4"):
    t = np.arange(t0, t_end + h, h)
    y = np.zeros(len(t))
    y[0] = y0
    
    for i in range(len(t)-1):
        if method == "euler":
            y[i+1] = y[i] + h * func(t[i], y[i])
            
        elif method == "heun":
            k1 = func(t[i], y[i])
            k2 = func(t[i] + h, y[i] + h * k1)
            y[i+1] = y[i] + (h/2) * (k1 + k2)
            
        elif method == "rk4":
            k1 = func(t[i], y[i])
            k2 = func(t[i] + h/2, y[i] + h/2 * k1)
            k3 = func(t[i] + h/2, y[i] + h/2 * k2)
            k4 = func(t[i] + h, y[i] + h * k3)
            y[i+1] = y[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
            
        elif method == "adams-bashforth":
            # Use RK4 to start the first step
            if i == 0:
                k1 = func(t[i], y[i])
                k2 = func(t[i] + h/2, y[i] + h/2 * k1)
                k3 = func(t[i] + h/2, y[i] + h/2 * k2)
                k4 = func(t[i] + h, y[i] + h * k3)
                y[i+1] = y[i] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
            else:
                f_n   = func(t[i], y[i])
                f_nm1 = func(t[i-1], y[i-1])
                y[i+1] = y[i] + (h/2) * (3*f_n - f_nm1)
        else:
            raise ValueError("Unknown method")
    
    return t, y
```

```python
import matplotlib.pyplot as plt

# Test problem
lam = 1.0
f = lambda t, y: -lam*y
y_exact = lambda t: np.exp(-lam*t)

t0, t_end, y0, h = 0, 5, 1, 0.1

methods = ["euler", "heun", "adams-bashforth", "rk4"]
results = {}

for method in methods:
    t, y = ode_solver(f, y0, t0, t_end, h, method)
    results[method] = y

t_fine = np.linspace(t0, t_end, 500)
plt.plot(t_fine, y_exact(t_fine), 'k-', label="Exact Solution")
plt.plot(t, results["euler"], 'r--', label="Euler")
plt.plot(t, results["heun"], 'g-.', label="Heun")
plt.plot(t, results["adams-bashforth"], 'm:', label="Adams–Bashforth")
plt.plot(t, results["rk4"], 'b-', label="RK4")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparison of Numerical ODE Methods")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute max error for each method
h_values = [0.5, 0.2, 0.1, 0.05, 0.01]
errors = {m: [] for m in methods}

for h in h_values:
    t_eval = np.arange(t0, t_end + h, h)
    exact = y_exact(t_eval)
    for method in methods:
        t, y = ode_solver(f, y0, t0, t_end, h, method)
        errors[method].append(np.max(np.abs(y - exact)))

# Print errors table
print("Step size | Euler   | Heun    | Adams-Bashforth | RK4")
for i, h in enumerate(h_values):
    print(f"{h:<9} | {errors['euler'][i]:<7.4f} | {errors['heun'][i]:<7.4f} | {errors['adams-bashforth'][i]:<15.4f} | {errors['rk4'][i]:<7.4f}")
```

## Conclusion

- RK4 is the most accurate method among the four for the same step size.  
- Heun and Adams–Bashforth are intermediate in accuracy.  
- Euler is the simplest but least accurate.  
- Error analysis confirms theoretical orders of accuracy:
  - Euler: $O(h)$
  - Heun: $O(h^2)$
  - Adams–Bashforth: $O(h^2)$
  - RK4: $O(h^4)$
- This notebook provides a clear comparison and can serve as a reference for other ODE problems.
