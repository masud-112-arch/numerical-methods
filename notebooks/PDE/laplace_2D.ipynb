# 2D Laplace Equation – Finite Difference Method

We solve the 2D Laplace equation:

$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0, \quad (x,y) \in \Omega
$$

- $u(x,y)$ = steady-state temperature or potential  
- Boundary conditions (Dirichlet) are specified on all edges.

---

## Finite Difference Discretization

Using central differences on a uniform grid:

$$
u_{i,j} = \frac{1}{4} \big(u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1}\big)
$$

- Iterate until convergence (Gauss–Seidel or Jacobi method).  
- This is steady-state diffusion, equivalent to Laplace’s equation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Lx, Ly = 1.0, 1.0
Nx, Ny = 20, 20
dx, dy = Lx/Nx, Ly/Ny
tolerance = 1e-5
max_iter = 5000

# Initialize grid
u = np.zeros((Nx+1, Ny+1))

# Boundary conditions
u[0, :] = 0        # left edge
u[-1, :] = 0       # right edge
u[:, 0] = 0        # bottom edge
u[:, -1] = 100     # top edge (hot boundary)

# Iterative solver (Jacobi method)
for it in range(max_iter):
    u_new = u.copy()
    u_new[1:-1, 1:-1] = 0.25*(u[2:, 1:-1] + u[:-2, 1:-1] + u[1:-1, 2:] + u[1:-1, :-2])
    
    error = np.max(np.abs(u_new - u))
    u[:] = u_new[:]
    
    if error < tolerance:
        print(f"Converged after {it} iterations")
        break

# Plotting the solution
X, Y = np.meshgrid(np.linspace(0, Lx, Nx+1), np.linspace(0, Ly, Ny+1))
plt.figure(figsize=(7,6))
cp = plt.contourf(X, Y, u.T, 20, cmap='hot')
plt.colorbar(cp)
plt.title("2D Laplace Equation – Steady-State Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Conclusion

- The 2D Laplace equation represents steady-state heat or potential distribution.  
- Central difference discretization and iterative solver (Jacobi method) work well for simple domains.  
- Boundary conditions strongly affect the solution.  
- Contour plots provide a clear visualization of steady-state temperature.  
- This notebook lays the foundation for 2D CFD simulations and further extensions to Poisson or Navier–Stokes equations.
